````python
Ya = np.sin(X)
Yb = np.cos(X)
plt.plot(X, Ya)
plt.plot(X, Yb)
plt.show()
````

````python
def plot_slope(X, Y):
    Xs = X[1:] - X[:-1]
    Ys = Y[1:] - Y[:-1]
    plt.plot(X[1:], Ys / Xs)
X = np.linspace(-3, 3, 100)
Y = np.exp(-X ** 2)
plt.plot(X, Y,'r')
plot_slope(X, Y)
plt.show()
````

````python
X = np.linspace(0, 2 * np.pi, 100)
Ya = np.sin(X)
Yb = np.sin(X)
Yc = np.sin(X)
Ya[Ya[:]>0]=0
Yc[Yc[:]<0]=0
plt.plot(X, Ya)
plt.plot(X, Yb)
plt.plot(X, Yc)
plt.plot(X, Yb,'k')
plt.plot(X, Ya,'k--')
plt.plot(X, Yc,'k--')
plt.plot(X, Yb,'k')
plt.fill_between(X,Ya , Yb, color='red', alpha='0.5')
plt.fill_between(X,Yc , Yb, color='blue', alpha='0.5')

#plt.grid()
#plt.grid(b=True, which='major', color='#666666', linestyle='-.')
plt.grid(b=True, which='major', color='#999999', linestyle='-', alpha=0.2)
plt.savefig('questions1.png')
````

````python
mm=[]
counter=0
for i,j,k,m,z in zip(A,B,C,SumA,range(len(A))):
    if i+j+k!=m:
        mm.append((z,m))
        counter +=1
        print(f'There is an error {i}+{j}+{k} not equal to {m} at index # {z}')
if len(mm)==0:
    print(f'No Error in this design')
else:
    print(f'There are {counter} errors in this design')    
````
