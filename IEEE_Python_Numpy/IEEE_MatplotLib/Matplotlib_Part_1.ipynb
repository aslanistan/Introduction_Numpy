````python
#Vs =10V, Rs=50 Ohms, RL=1-100Ohms
#VRL=(Vs)RL/(RS+RL)
#IRL=VRL/RL
Vs=10;
Rs=50;
RL=[]
for i in range(1,101):
    RL.append(i)
VRL=[]
IRL=[]
for i in RL:
    VRL.append(Vs*i/(Rs+i))
for i,j in zip(VRL,RL):
    IRL.append(i/j)
    

fig = plt.figure()
ax1 = fig.add_subplot(1, 1, 1)
ax2 = ax1.twinx()
ax1.plot(RL, VRL, 'b-')
ax2.plot(RL, IRL, 'r-')
ax1.set_ylabel('VRL', color='red')
ax2.set_ylabel('IRL', color='blue')
ax1.set_xlabel('RL (Ohms)')
````

````python
# Initialize figure and axes
fig = plt.figure(figsize=(8, 6))
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])

# Define spectral types
spectral_id = [1, 2, 3, 4, 5, 6, 7]
spectral_types = ['O', 'B', 'A', 'F', 'G', 'K', 'M']

# Plot the data
ax.plot(spectral_id, [4, 3, 2, 3, 4, 5, 4], 'ro')

# Set the limits
ax.set_xlim(1, 7.5)
ax.set_ylim(0., 10.)

# Set the custom ticks on the x-axis
ax.set_xticks(spectral_id)
ax.set_xticklabels(spectral_types)

# Set the axis labels
ax.set_xlabel("Spectral type")
ax.set_ylabel("Number of sources")
````

````python
import numpy as np
import matplotlib.pyplot as plt 
from matplotlib import gridspec

# generate some data
x = np.arange(0, 10, 0.2)
y = np.sin(x)

# plot it
fig = plt.figure(figsize=(8, 6)) 
gs = gridspec.GridSpec(2, 3, width_ratios=[3.5, 0.5,1]) 
ax0 = plt.subplot(gs[0])
ax0.plot(x, y)
ax1 = plt.subplot(gs[1])
ax1.plot(y, x)
ax1 = plt.subplot(gs[2])
ax1.plot(y, x)
ax1 = plt.subplot(gs[3])
ax1.plot(y, x)

plt.tight_layout()
plt.savefig('grid_figure.pdf')
````
